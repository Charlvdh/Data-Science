#### Introduction

Let's assume that we're working for an an e-learning company that offers courses on programming. Most of our courses are on web and mobile development, but we also cover many other domains, like data science, game development, etc. 

We want to promote our product and we'd like to invest some money in advertisement. Our goal in this project is to find out the two best markets to advertise our product in.

#### Approach

To reach our goal, we could have organized surveys for a couple of different markets to find out which would the best choices for advertising. This is very costly, however, and it's a good call to explore cheaper options first.

We instead searched existing data relevant for our purpose. One good candidate was the data from freeCodeCamp's 2017 New Coder Survey. freeCodeCamp is a free e-learning platform that offers courses on web development. Because they run a popular Medium publication (over 400,000 followers), their survey attracted new coders with varying interests (not only web development), which is ideal for the purpose of our analysis.

The survey data is publicly available in [this GitHub repository](https://github.com/freeCodeCamp/2017-new-coder-survey).

#### Further exploration

As we mentioned earlier, most of the courses we offer are on web and mobile development, but we also cover many other domains, like data science, game development, etc. For the purpose of our analysis, we want to answer questions about a population of new coders that are interested in the subjects we teach. We'd like to know:

- Where are these new coders located.
- What are the locations with the greatest number of new coders.
- How much money new coders are willing to spend on learning.

Before starting to analyze the sample data we have, we need to clarify whether it's representative for our population of interest and it has the right categories of people for our purpose.

Lets have a look at what subjects the survey participants are interested in and determine if this is representative of the population as well as if anyone is interested in our two focus areas (web and mobile development).

```python
interest_columns = []

for i in survey.columns:
    if len(i) > 11:
        if i[:11] == "JobInterest":
            interest_columns.append(i)
        
interest_columns
```

```python
interests_df = survey[interest_columns]

# Let's clean up the column names
new_columns = []
for c in interests_df.columns:
    new_columns.append(c[11:])
    
interests_df.columns = new_columns

percentages_interested_per_subject = (interests_df.describe().iloc[0] / interests_df.shape[0]) * 100

import matplotlib.pyplot as plt
%matplotlib inline


# Lets show the topics applicable to us () web and mobile development in green and the others in blue
colors = ["b","b","b","g","b","b","g","g","b","b","b","g"]

# The colors list needs to given in reverse order
colors.reverse()

percentages_interested_per_subject.plot.barh(color=colors)

plt.title("Percentage of total respondents interested in each topic", pad=10,
          fontdict={'fontsize': 14})
plt.xlabel("Percentage")
plt.ylabel("Interest areas")
```

Thus we can see that while mobile and web development are the topics that are generating the most interest from freeCodeCamp's followers, the survey respondents are interested in a broad range of topics ranging from UX (User experience) to DataEngr (Data engineering).

As our goal is to determine how best to advertise our web and mobile development ratated courses, this dataset is highly applicable as there are many respondents interested in these particular topics. By isolating only the people interested in our content, we can best refine our advertising strategy to generate the most sales.

#### Determining where the respondents are located

Now that we know that the sample has the right categories of people for our purpose, we can begin analyzing it. We can start with finding out where these new coders are located and what the densities are (how many coderes there are) for each location.

The data set provides information about the location of each participant at a country level. The `CountryCitizen` variable describes the country of origin for each participant, and the `CountryLive` variable describes what country each participants lives in (which may be different than the origin country).

For our analysis, we'll work with the `CountryLive` variable because we're interested where people actually live at the moment when we run the ads. In other words, we're interested where people are located, not where they were born.

Because the data set provides information at a country level, we can think of each country as an individual market. This means we can frame our goal as finding the two best countries to advertise in.

```python
survey["CountryLive"].value_counts(normalize=True).head(20) * 100
```

```python
survey["CountryLive"].value_counts().head(10)
```

As we can see from the above frequency tables, the United States of America is by far the largest in terms of number of respondents representing almost 38% of respondents or 5791 people.

The second largest population is india with 9.1% of respondents. This is followed by the United Kindom (4.9%), and Canada (4%). After this the next highest is only 2.37%.

Based on these results the USA and India seem prime candidates for advertising targets but up to this point we have ignored important factors such as which topics each of these counties is interested in. i.e. are they interested in mobile and web development.

We have also ignored possible competitors which target these countries.

Let's explore our first concern and determine the top two countries by number of respondents interested in mobile and we development.

```python
relevant_interests = ["JobInterestBackEnd","JobInterestFrontEnd","JobInterestFullStack","JobInterestMobile"]
ris = relevant_interests

relevant_interests_df = survey.loc[((survey[ris[0]]==1) 
                                    | (survey[ris[1]]==1) 
                                    | (survey[ris[2]]==1) 
                                    | (survey[ris[3]]==1))
                                   ,"CountryLive"]

relevant_interests_df.value_counts(normalize=True).head(10) * 100
```

This confirms that the US and India still seem to be the countries in which we sould focus our advertising prior to considering competition.

#### Analysis of potential customer's budgets

Previously, we found useful information about the location of new coders, and what are the countries where new coders live, but we need to go more in depth with our analysis before taking a decision. We can continue by figuring out how much money new coders are actually willing to spend on learning. Advertising within markets where most people are only willing to learn for free is extremely unlikely to be profitable for us.

The MoneyForLearning column describes in American dollars the amount of money spent by participants from the moment they started coding until the moment they completed the survey. Our company sells subscriptions at a price of $59 per month, and for this reason we're interested in finding out how much money each student spends per month.

It also seems like a good idea to narrow down our analysis to only four countries: the US, India, the United Kingdom, and Canada. Two reasons for this decision are:

- These are the countries having the highest absolute frequencies in our sample, which means we have a decent amount of data for each.
- Our courses are written in English, and English is an official language in all these four countries. The more people that know English, the better our chances to target the right people with our ads.

Let's determine the monthly education spend for each student in our sample who are interested in one of our topics

```python
respondents_4_countries = survey.loc[((survey[ris[0]]==1) 
                                    | (survey[ris[1]]==1) 
                                    | (survey[ris[2]]==1) 
                                    | (survey[ris[3]]==1))]

respondents_4_countries = respondents_4_countries.loc[(survey["CountryLive"].isin(["United States of America", "India", "United Kingdom", "Canada"]))]

# We need to determine if any respondents have a `MonthsProgramming` value of 0 as this will give us
# infinite monthly education spend values
respondents_4_countries.loc[(respondents_4_countries["MonthsProgramming"] == 0), "MonthsProgramming"].count()
```

In order to rectify this problem of infinite values, let's change all `MonthsProgramming` values of 0 to values of 1

```python
def convert_0_to_1(num):
    if num == 0:
        return 1
    else:
        return num
    
respondents_4_countries["MonthsProgramming"] = respondents_4_countries["MonthsProgramming"].apply(convert_0_to_1)

respondents_4_countries["MonthsProgramming"].describe()
```

```python
respondents_4_countries["monthly_spend"] = respondents_4_countries["MoneyForLearning"] / respondents_4_countries["MonthsProgramming"]
```

Let's also determine if there are any NULL values in the `monthly_spend` column and drop these rows

```python
respondents_4_countries["monthly_spend"].isnull().sum()
```

Let's drop all of the rows that have NULL for the new column.

```python
respondents_4_countries.dropna(subset=["monthly_spend"], inplace=True)

respondents_4_countries["monthly_spend"].isnull().sum()
```

Now lets calculate the mean spend per country

```python
respondents_4_countries.groupby(["CountryLive"]).mean()["monthly_spend"]
```

The results for the United Kingdom and Canada are surprisingly low relative to the values we see for India. If we considered a few socio-economical metrics (like GDP per capita), we'd intuitively expect people in the UK and Canada to spend more on learning than people in India.

It might be that we don't have have enough representative data for the United Kingdom, Canada, and India, or we have some outliers (maybe coming from wrong survey answers) making the mean too big for India, or too low for the UK and Canada. Or it might be that the results are correct.

```python
respondents_4_countries["monthly_spend"].describe()
```

From the above we can see that the max monthly spend is $80,000. This seems to be an outlier.

#### Extreme outliers

Let's determine if there are any extreme outliers that need to be eliminated

```python
# Let's first change the column names for United States of America and United Kindom to US and UK
def shorten_country_names(name):
    if name == "United States of America":
        return "US"
    elif name == "United Kingdom":
        return "UK"
    else:
        return name
    
respondents_4_countries["CountryLive"] = respondents_4_countries["CountryLive"].apply(shorten_country_names)
        

import seaborn as sns

sns.boxplot(x="CountryLive", y="monthly_spend", data=respondents_4_countries)
sns.catplot(x="CountryLive", y="monthly_spend", data=respondents_4_countries)
```

Let's remove all values over 10000 as it is extremely unlikely that any of these values are valid

```python
good_data = respondents_4_countries.loc[(respondents_4_countries["monthly_spend"] < 10000)]

sns.boxplot(x="CountryLive", y="monthly_spend", data=good_data)
sns.catplot(x="CountryLive", y="monthly_spend", data=good_data)
```

```python
good_data.groupby(["CountryLive"]).mean()["monthly_spend"]
```

Let's have a look at the extreme values in India. I.e. the values over \$1000 and try to determine why the values are so extreme

```python
india_over_1000 = good_data.loc[((good_data["monthly_spend"] > 1000) 
                                             & (good_data["CountryLive"] == "India"))]

india_over_1000
```

Conclusion:

- Only one of these respondents attended a bootcamp and so that does not seem to be the factor driving the high spending. 
- All respondents are under 27 years of age.
- Most are interested in Full Stack development.
- Most have a relativley low `MonthsProgramming` value.
- The highest `MoneyForLearning` value is 100,000 and the lowest is 2,000

The actual question of the survey was "Aside from university tuition, about how much money have you spent on learning to code so far (in US dollars)?", so they might have misunderstood and thought university tuition is included.

It seems safer to remove these values.

```python
good_data.drop(india_over_1000.index, inplace=True)
```

```python
good_data.groupby(["CountryLive"]).mean()["monthly_spend"]
```

There also seem to be a few extreme outliers in the US. Let's have a closer look at these.

```python
us_over_5000 = good_data.loc[(good_data["CountryLive"] == "US") & (good_data["monthly_spend"] > 5000)]

us_over_5000
```

From analysis of the above it seems that all of these respondents attended a bootcamp and all have been programming for 3 months or less.

Let's have a look at the fees for each of the bootcamps and compare this to the `MoneyForLearning` values:

- [The Coding Boot Camp at UCLA Extension](https://www.switchup.org/bootcamps/ucla-extension-boot-camps) : 10,995 for 24 weeks (1,833pm) - `MoneyForLearning` = 8000
- [The Iron Yard](https://www.newsobserver.com/news/business/article162998653.html) : 13,900 for 12 weeks (4,633pm) - `MoneyForLearning` = 13000
- [Hackbright Academy](https://hackbrightacademy.com/payment-plans/) : 16,985 for 12 weeks (5,661pm) - `MoneyForLearning` = 17000
- [Codeup](https://www.bloc.io/learn-to-code/bc/codeup) : 16,000 for 16 weeks (4,000pm) - `MoneyForLearning` = 20000
- [Grand Circus](https://www.switchup.org/bootcamps/grand-circus) : 9,750 for 12 weeks (3,250pm) - `MoneyForLearning` = 8000


Thus these values are likely due to bootcamp fees and should not be excluded from the sample.

Let's also look at the extreme outlier for Canada which looks to be at about \$5000

```python
canada_over_4000 = good_data.loc[(good_data["CountryLive"] == "Canada") & (good_data["monthly_spend"] > 4000)]

canada_over_4000
```

This respondent took part in the [Bloc.io](https://www.bloc.io/web-developer-track) Bootcamp program which costs \$7,500 for the 8 month program.

This explains the `MoneyForLearning` value of \$10,000 and when combined with the `MonthsProgramming` value of 2, explains the high `monthly_spend` value of \$5,000.

This is not an invalid data point as someone who has attended a bootcamp could well be interested in keeping up to date with new technoogies through use of our courses.

#### Final sample

For our final sample which we will use to generate mean spend per country, we will remove all respondents with `monthly_spend` values of over \$3,000 as these values, although valid, stronghly influence our mean spends per country and show a biased picture of the average respondent who would take an online course.

We can also see from the previous sns.catplot that there are relativly few datapoints over \$3000.

```python
final_sample = good_data.loc[(good_data["monthly_spend"] < 3000)]

final_sample.groupby(["CountryLive"]).mean()["monthly_spend"]
```

#### Choosing the best two markets

As per the sample above and considering that the monthly subscription price of our product is \$59, it seems that the **US** and **Canada** would be the most advantageous markets in which to advertise.
